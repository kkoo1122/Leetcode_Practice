```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        
        
        n1 , n2 = 0 , 0
        
        while l1 or l2:
            if l1:
                n1 = n1*10 + l1.val
                l1 = l1.next
            if l2:
                n2 = n2*10 +l2.val
                l2 = l2.next
            
        num = str(n1+n2)
        
        dummy= node = ListNode(0)
        
        for i in num:
            node.next = ListNode(int(i))
            node = node.next
        
        
        return dummy.next
                
```
