### Thought

The most important part is to think about the **(remains of the k)/(length of the list)**. And we want to move the **steps = length of the list - remains (steps = cnt - k%cnt)**<br>

**===Code:===**

    <p>p1.next=head</p>
    <p>head=end.next</p>
    <p>end.next=None</p>

is crucial method to change the some node of the list in the middle to the head.


    

```python
def rotateRight(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: ListNode
        """
        if not head or not head.next: return head
        
        p1=end=head
        cnt=1
        
        while p1.next:
            p1=p1.next
            cnt+=1
        
        step=cnt-k%cnt #steps from the begining
        
        if step==0:
            return head
        
        while step>1:
            end=end.next
            step-=1
        
        p1.next=head
        head=end.next
        end.next=None   
        
        
        
        return head
```
