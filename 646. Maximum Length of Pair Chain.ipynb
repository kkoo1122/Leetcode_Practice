```python
def findLongestChain(self, pairs):
        cur, ans = float('-inf'), 0
        for x, y in sorted(pairs, key = lambda z:z[1]):
            if cur < x:                                
                cur = y
                ans += 1
        return ans
L=[[-6,9],[1,6],[8,10],[-1,4],[-6,-2],[-9,8],[-5,3],[0,3]]

findLongestChain(0,L)
```
