```python
def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """
        if numRows==1: return s
        
        ZigZag=['' for _ in range(numRows)]
        
        line=numRows-1
        zz=0
        flag=1
        for index in s:
            if flag==1:
                ZigZag[zz]+=index
                zz+=1
                if zz==line:
                    flag=-1
            elif flag==-1:
                ZigZag[zz]+=index
                zz-=1
                if zz==0:
                    flag=1
        

        res=''
        for j in range(len(ZigZag)):
            res+=ZigZag[j]
            
        return res
```
