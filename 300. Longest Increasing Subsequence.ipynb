```python
nums=[10,9,2,5,3,7,101,18]

dp=[1]*(len(nums))
for i in range(1,len(nums)):
    for j in range(i):
        if nums[i] > nums[j]:
            dp[i]=max(dp[i],dp[j]+1)
    
res=max(dp)
print(res)
```

```python

def num_ways_X(n,X):
    if n==0: return 1
    total=0
    for i in X:
        if n-i >=0: total += num_ways_X(n-i,X)
    return total

def numways_X_bootomup(n,X):
    if n == 0: return 1
    nums = [0 for i in range(n+1)]
    
    nums[0]=1
    for i in range(1,n+1):
        total = 0
        for j in X:
            if i-j >=0 :
                total += nums[i-j]
        nums[i]=total
    return nums[n]
    
print(num_ways_X(6,[1,2]))
print(numways_X_bootomup(6,[1,2]))
```
