```python
def rob(nums):
    a,b,c=0,0,0
    for h in range(2,len(nums)+1):
        c =min(a+nums[h-2],b+nums[h-1])
        a,b=b,c
        h+=1
    return sum(nums)-b


rob([1,2,3,4,5,6])     
        
```

```python
def rob(nums):
    a,b,res=0,0,0
    for num in nums:
        res=max(a+num,b)
        a,b=b,res
    
    return res

rob([1,2,3,4,5,6])  
```

```python
#Recursion + Memorize
#Time Complexity:O(n)
#Space Complexity:O(n)

rob([2,1,1,2]) 
```

```python
#DP
def rob(nums):
    dp=[0]*len(nums)
    dp[0],dp[1]=nums[0],nums[1]
    
    for n in range(1,len(nums)):
        
        dp[n]=max(dp[n-2]+nums[n],dp[n-1]) # not two consecutive number
    
    return dp[len(nums)-1]


rob([1,2,3,4]) 
```
