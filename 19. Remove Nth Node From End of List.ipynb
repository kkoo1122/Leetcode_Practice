### Thoughts
I'm using two pointers, p1 and p2 to point the list. p1 is going forward by "n" from p2.<br>
**Speacial condition** like: [1,2,3,4,5], n=5. What we have to do is eliminating the first node. Therefore, we should return head.next.

The while loop is just making p1 keep going until it hits the end of the list. We want to make **p2.next=p2.next.next** to **eliminate the number we want**. Therefore, **before the while loop, p1=p1.next** is necessary to make the distance between p1 and p2 become n+1

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def removeNthFromEnd(self, head, n):
        """
        :type head: ListNode
        :type n: int
        :rtype: ListNode
        """
        if not head.next: return []        
        p1=p2=head
        for _ in range(n):
            p1=p1.next    
        if not p1: 
            return head.next #for the case like [1,2] and n=2
        p1=p1.next      
        while p1:
            p1=p1.next
            p2=p2.next
        p2.next=p2.next.next
        
        return head
        
        
        
```
