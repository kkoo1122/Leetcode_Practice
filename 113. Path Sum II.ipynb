```python
def pathSum(self, root, target):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: List[List[int]]
        """   
        if not root: return []
        if not root.left and not root.right:
            if target-root.val ==0: return [[target]]
            else: return []
        
        return  self.dfs(root,target,[],[])
    
    def dfs(self,root,target,L,res):
        if not root.left and not root.right:
            if target-root.val==0:
                res.append(L+[root.val])
                return
        if root.right:    
            self.dfs(root.right,target-root.val,L+[root.val],res)
        if root.left:
            self.dfs(root.left,target-root.val,L+[root.val],res)
        
        return res
```
